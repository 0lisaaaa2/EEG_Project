```python
import mne
import numpy as np
import matplotlib.pyplot as plt
from mne_bids import BIDSPath, read_raw_bids
```

```python
# Change these
subject = "039"
task = "regfront"   # or regperp

bids_root = "ds005841-download"  # path to your BIDS dataset

bids_path = BIDSPath(
    subject=subject,
    task=task,
    datatype="eeg",
    suffix="eeg",
    extension=".bdf",
    root=bids_root
)

raw = read_raw_bids(bids_path)
raw.load_data()

print(raw)
```

```python
print("All channel names:")
print(raw.ch_names)

# Try common trigger channel names
possible_status_names = ["Status", "STI 014", "Trigger", "TRIG", "STIM"]

status_ch = None
for ch in possible_status_names:
    if ch in raw.ch_names:
        status_ch = ch
        break

if status_ch is None:
    raise RuntimeError("❌ No status channel found!")

print(f"\n✅ Using status channel: {status_ch}")
```

```python
status_data = raw.copy().pick_channels([status_ch]).get_data()[0]

print("Unique Status values:")
unique_vals = np.unique(status_data.astype(int))
print(unique_vals)
print(f"Total unique values: {len(unique_vals)}")

nonzero = np.sum(status_data != 0)
print(f"Number of non-zero samples: {nonzero}")
```
